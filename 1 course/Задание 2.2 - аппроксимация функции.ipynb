# Задание 2.2 - аппроксимация функции

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
import math
import scipy
#interpolate - приближение графика
```

```python
def f(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5 * math.exp(-x / 2.0)
```

```python
x = np.arange(1, 15, 0.1)
y = []
for i in x: y.append(f(i))
plt.plot(x, y)
plt.show()
```

## 1

```python
#Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A и свободный вектор b) для многочлена 
#первой степени, который должен совпадать с функцией f в точках 1 и 15. Решите данную систему с помощью функции 
#scipy.linalg.solve. Нарисуйте функцию f и полученный многочлен. Хорошо ли он приближает исходную функцию?
A = [[1, 1], [1, 15]]
b = [f(1), f(15)]
w0, w1 = scipy.linalg.solve(A, b)
```

```python
def g_1(x):
    return w0 + w1 * x
x1 = np.arange(1, 15, 0.1)
y1 = []
for i in x1: y1.append(g_1(i))
plt.plot(x, y, 'o', x1, y1, '-')
plt.show()
```

## 2

```python
#Повторите те же шаги для многочлена второй степени, который совпадает с функцией f в точках 1, 8 и 15. 
#Улучшилось ли качество аппроксимации?
A = [[1, 1, 1], [1, 8, 64], [1, 15, 225]]
b = [f(1), f(8), f(15)]
w0, w1, w2 = scipy.linalg.solve(A, b)
```

```python
def g_2(x):
    return w0 + w1 * x + w2*(x**2)
x2 = np.arange(1, 15, 0.1)
y2 = []
for i in x2: y2.append(g_2(i))
plt.plot(x, y, 'o', x2, y2, '-')
plt.show()
```

## 3

```python
#Повторите те же шаги для многочлена третьей степени, который совпадает с функцией f в точках 1, 4, 10 и 15. 
#Коэффициенты данного многочлена (четыре числа в следующем порядке: w_0, w_1, w_2, w_3) являются ответом на задачу.
A = [[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 15**3]]
b = [f(1), f(4), f(10), f(15)]
w0, w1, w2, w3 = scipy.linalg.solve(A, b)
```

```python
def g_3(x):
    return w0 + w1 * x + w2*(x**2) + w3*(x**3)
x3 = np.arange(1, 15, 0.1)
y3 = []
for i in x3: y3.append(g_3(i))
plt.plot(x, y, 'o', x3, y3, '-')
plt.show()
```
