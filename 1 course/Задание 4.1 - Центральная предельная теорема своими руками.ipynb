В качестве начального распределения я выбрал равномерное распределение на отрезке $[0, 30]$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

Сгенерируем из неё выборку, объема 1000:

```python
uniform_rv = sts.uniform(0, 30)
sample = uniform_rv.rvs(1000)
```

Сгенерируем из него выборку объёма 1000, построим гистограмму выборки и нарисуем поверх неё теоретическую плотность распределения моей случайной величины

```python
plt.hist(sample, normed=True, label = 'hystogram')

x = np.linspace(-15,45,1000)
pdf = uniform_rv.pdf(x)

plt.plot(x, pdf, label = 'Theoretical')

plt.legend(loc='upper right')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Сгенерируйте 1000 выборок объёма n = 2 и постройте гистограммы распределений их выборочных средних.

```python
middle_5 = []
for i in range(0, 1000):
    sample_5 = np.random.choice(sample, 5)
    sum = 0.0
    for j in range (0, 5):
        sum += sample_5[j]
    middle_5.append(sum / 5.0)
```

```python
plt.hist(middle_5, normed=True, label = 'n = 5')
plt.legend(loc='upper right')
plt.ylabel('$\overline{f}(x)$')
plt.xlabel('$x$')
plt.show()
```

Для равномерное распределение на отрезке $[0,30]$: $\mu = \frac{30 - 0}{2} = 15$ и $\sigma^2 = \frac{(30-0)^2}{12} = 75$

Будем использовать формулу, что $\mu_n = \mu$ и $\sigma_n^2 = \frac{\sigma^2}{n}$

Поверх каждой гистограммы нарисуем плотность соответствующего нормального распределения

## n = 5

```python
mu = 15.0
sigma = math.sqrt(75/5.0)
norm_rv = sts.norm(loc=mu, scale=sigma)
```

```python
plt.hist(middle_5, normed=True, label = 'hystogram')

x = np.arange(0, 30, 0.01)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label = 'theoretical')

plt.legend(loc='upper right')
plt.ylabel('$\overline{f}(x)$')
plt.xlabel('$x$')
plt.show()
```
